```python
df['Review Rating']= df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(x=df['Review Rating'])
plt.show()
```

```python
df.columns = df.columns.str.lower 
df.columns = df.columns.str.replace(' ','_') 
```

```python
df.columns
```

```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create a columns age group
labels=['Young Adult','Adult','Middle-aged','Senior']
#It divides (or bins) a continuous variable into equal-sized quantile groups â€” 
#meaning each group contains approximately the same number of observations.
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(20)
```

```python
# create column purchase_frequency_days
frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 months':90
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```
